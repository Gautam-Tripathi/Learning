```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("diabetes.csv")
df.head()
```

# Visualize the distribution of glucose levels in the using a boxplot.

```python
df['Glucose'].plot(kind = 'box',grid = True)
```

# Plot the distribution of the Body Mass Index (BMI) variable using a histogram.

```python
df['BMI'].plot(kind='hist')
```

# Create a distribution plot for the skin thickness variable.

```python
df['SkinThickness'].plot(kind='density')
df['SkinThickness'].plot(kind='hist',density = True)
```

# Create a scatterplot showing the relationship between the age and glucose levels.

```python
ax = df.plot(['Age'],['Glucose'],kind='scatter')
```

# Visualize the relationship between the BMI and insulin variables using a bubble plot.

```python
fig = plt.scatter(x= df['Insulin'],y =df['BMI'],s = df['BMI']*2,alpha = 0.2)
```

# Create a bar chart to show the frequency of diabetes cases (positive/negative).

```python
cases = df['Outcome'].unique()
counts = df['Outcome'].value_counts()
plt.bar(cases,counts,label = df['Outcome'].unique())
```

# Create a line chart to show the change in blood pressure levels over time for a selected group of patients.

```python
fig = plt.figure()
plt.plot(df['BloodPressure'],df['Age'])
```

# Create a pie chart to show the proportion of positive and negative diabetes cases.

```python
fig, ax = plt.subplots()
ax.pie(df['Outcome'].value_counts().values,df['Outcome'].value_counts().index, autopct='%1.1f%%',startangle=90)
```
