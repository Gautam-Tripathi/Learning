```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('diabetes.csv')
df.head()
```

# 1. Histogram 
## Visualize the distribution of Body Mass Index(BMI) among people in the Pima Indian Community.

```python
sns.histplot(df['BMI'], kde=False)
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(df['BMI'],kde = True,color='orange')
#KDE = Kernel Densiity estimate : represent data using a continuous probability density curve
plt.title('Distribution of BMI')
plt.xlabel('BMI')
plt.ylabel('Count')
plt.show()
```

# 2. Density Plot
## Estimate the probability density function of blood pressure among people in the Pima Indian Comunity with Diabetes

```python
sns.kdeplot(df[df['Outcome']==1]['BloodPressure'],label = 'Diabetes')
plt.title('Density Plot of Blood Pressure by Diabetes Status')
plt.xlabel('Blood Pressure')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
sns.kdeplot(df[df['Outcome']==1]['BloodPressure'],label = 'Diabetes')
sns.kdeplot(df[df['Outcome']==0]['BloodPressure'],label = 'No Diabetes')
plt.title('Density Plot of Blood Pressure by Diabetes Status')
plt.xlabel('Blood Pressure')
plt.ylabel('Density')
plt.legend()
plt.show()
```

## Let's plot density plot for age among people in the Pima Indian community with and without a history of gestational diabetes to analyze whether is any significant difference in distribution of age

```python
df2 = df[df['Age']>30]
sns.kdeplot(df2[df2['Outcome']==0]['Age'], label='No Gestational Diabetes')
sns.kdeplot(df2[df2['Outcome']==1]['Age'], label='Gestational Diabetes')
plt.title('Density Plot of Age by Gestational Diabetes History')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

# 3. Box Plot
## Let us plot a box plot illustrating distribution of glucose for people with diabetes

```python
fig,ax = plt.subplots(figsize=(8,8))
sns.boxplot(data= df,x = 'Outcome',y= 'Glucose')
plt.title('Distribution of Glucose by Diabetes Outcome')
plt.xlabel('Outcome')
plt.ylabel('Glucose')
plt.show()
```

# 4. Violin Plot
This combines a Box plot and kde plot.It displays the distribution of data and provides insights into its shape, symmetry, and outliers.
## Combine the information from the box plot and density plot for age among people in the Pima Indian community with and without a history of gestational diabetes to investigate potential long-term health consequences and potential risk factors for future pregnancies.

```python
sns.violinplot(data=df[df['Age']>30], x= 'Outcome',y='Age')
plt.title('Distribution of Age by Gestational Diabetes History')
plt.xlabel('Outcome')
plt.ylabel('Age')
plt.show()
```

#  Scatter plot
## Create a scatter plot of Age and Blood Pressure for people with diabetes to analyze impact of age on blood pressure.

```python
sns.scatterplot(data=df[df['Outcome']==1],x='Age',y='BloodPressure')
plt.title('Scatter Plot of Age and BloodPressure for People with Diabetes')
plt.xlabel('Age')
plt.ylabel('Blood Pressure')
plt.show()
```

# Joint Plot
## Explore the relationship between age and BMI among people in the Pima Indian community and also, plot the individual distribution.

```python
sns.jointplot(data =df , x='Age',y='BMI')
plt.title('')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.show()
```

# Pair plot
## We can explore how "Insulin", "SkinThickness", "Age", and "BMI" relate to each other in the pima Indians diabetes dataset using pair plot and analyze how these relationships differ between individuals with and without diabetes.

```python
sns.pairplot(df, vars=['Insulin', 'SkinThickness', 'Age', 'BMI'], hue='Outcome')
plt.show()
```

# Heat Map
## Explore the correlation between variables in the Pima Indians Diabetes Dataset using a heat map.

```python
sns.heatmap(data= df.corr(),annot =True,cmap='coolwarm')
plt.show()
```
