```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("AirPassengersDataset.csv")
print(df.shape)
df.head()
```

## Read the First six months of (Jan-Jun) data into the first_half_df dataframe. And Read the next six months of data into the second_half_df dataframe.

```python
first_half_df = df.loc[:,'Jan':'Jun']
second_half_df = df.loc[:,'Jul':'Dec']
display(first_half_df)
display(second_half_df)
```

# Create a new dataframe air_df from the dataframes first_half_df and second_half_df by inner join on their respective indices.



```python
air_df = first_half_df.merge(right=second_half_df,how='inner',)
```
