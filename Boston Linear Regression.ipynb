```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Datasets/boston_housing.csv")
df.head()
```

## Consider the column, ‘RM’, as predictor, and ‘MEDV’ as the target variable 
## Visualize the association between the predictor and the target using scatter plot.

```python
sns.scatterplot(data= df,x= 'RM',y='MEDV')
plt.title("Visualization of Column predictor and Target")
plt.xlabel("RM")
plt.ylabel("MEDV")
plt.show()
```

## Split the data into train and test datasets, in the ratio of 67:33.

```python
X_train, X_test, y_train, y_test = train_test_split( df[['RM']], df['MEDV'], test_size=0.33)
```

## Build a Linear Regression model using training dataset, to predict the target variable.
## Observe the coefficient and intercept values for the model.



```python
model = LinearRegression()
model.fit(X_train, y_train)

print("Intercept",model.intercept_)
print("Coefficient", model.coef_)
```

## Evaluate the model using mean squared error values and R-squared values on the training and the testing datasets.

```python
training_data_score = model.score(X_train,y_train)
print(training_data_score)
test_data_score = model.score(X_test,y_test)
print(test_data_score)
print(model.score(X_test,model.predict(X_test)))
```

```python
plt.scatter(X_test, y_test, color="black")
plt.plot(X_test, model.predict(X_test), color="blue", linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
